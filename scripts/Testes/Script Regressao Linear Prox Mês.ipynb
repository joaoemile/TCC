```python
# Bibliotecas
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#base de dados
file_path = 'C:\\Users\\JoãoEmileXimenesMuri\\Documents\\pessoal\\TCC\\Bases TCC\\Dados_TCC_2014_2024_Normalizados_TabUnica.xlsx'
data = pd.read_excel(file_path, sheet_name='Sheet1')
```

```python
# Visualização inicial dos dados
print(data.head())
```

```python

# Transformar a coluna de data em datetime
data['Data Referência'] = pd.to_datetime(data['Data Referência'])
```

```python
# Criar a variável-alvo (target): Valor IPCA do próximo mês
# Criar uma nova coluna deslocada para representar o IPCA do próximo mês
data['IPCA Próximo Mês'] = data['Valor IPCA'].shift(-1)
```

```python
# Remover linhas com valores ausentes (causados pelo deslocamento)
data = data.dropna()
```

```python
# Definir as features (todas as variáveis exceto Data e a variável-alvo)
X = data.drop(columns=['Data Referência', 'IPCA Próximo Mês'])  # Variáveis explicativas
y = data['IPCA Próximo Mês']  # Variável alvo
```

```python
# Verificar as dimensões dos dados após a preparação
print(f"Dimensões das features (X): {X.shape}")
print(f"Dimensão do target (y): {y.shape}")
```

```python

# Divisão 80/20 mantendo ordem temporal
split_index = int(len(X) * 0.8)
X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]
```

```python

# Inicializar o modelo de regressão linear
model = LinearRegression()
```

```python
# Treinar o modelo com os dados de treinamento
model.fit(X_train, y_train)
```

```python

# Fazer previsões no conjunto de teste
y_pred = model.predict(X_test)
```

```python
# Calcular métricas de avaliação
mse = mean_squared_error(y_test, y_pred)  # Erro médio quadrático
r2 = r2_score(y_test, y_pred)  # Coeficiente de determinação
```

```python
print("\nMétricas de Avaliação:")
print(f"Erro Médio Quadrático (MSE): {mse:.4f}")
print(f"Coeficiente de Determinação (R²): {r2:.4f}")
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error

mae = mean_absolute_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)

print("Mean Absolute Error:", mae)
print("Mean Absolute Percentage Error:", mape)
```
